<h1>Data Pre-Processing</h1>

<h3>Dealing with Missing Data</h3>

```python
df.replace("?", np.nan, inplace = True)
```

```python
#evaluating missing values
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
#Replace by Mean
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

avg_bore=df['bore'].astype('float').mean(axis=0)
df["bore"].replace(np.nan, avg_bore, inplace=True)

avg_stroke=df['stroke'].astype('float').mean(axis=0)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
#Replace by Frequency
max_element = df['num-of-doors'].value_counts().idxmax()
df["num-of-doors"].replace(np.nan, max_element, inplace=True)
```

```python
#Dropping Rows
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

<h3>Data Formatting</h3>

```python
df.dtypes
```

<b>Converting data type to correct format</b>

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
df['city-L/100km'] = 235/df["city-mpg"]
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]
# rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)
```

<h3>Data Normalization</h3>

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test.head(2)
```

```python
df_gptest
```

```python
grouped_test.get_group('4wd')['price']
```

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test.get_group('fwd')['price'], grouped_test.get_group('rwd')['price'], grouped_test.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val) 
```

<p><b>Checking the correlation between each group</b></p>

```python
# fwd and rwd
f_val, p_val = stats.f_oneway(grouped_test.get_group('fwd')['price'], grouped_test.get_group('rwd')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```python
# 4wd and rwd
f_val, p_val = stats.f_oneway(grouped_test.get_group('4wd')['price'], grouped_test.get_group('rwd')['price'])
print( "ANOVA results: F=", f_val, ", P =", p_val) 
```

```python
# 4wd and fwd
f_val, p_val = stats.f_oneway(grouped_test.get_group('4wd')['price'], grouped_test.get_group('fwd')['price'])
print("ANOVA results: F=", f_val, ", P =", p_val) 
```
